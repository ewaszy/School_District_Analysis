```python
# Add the Pandas dependency.
import pandas as pd
```

```python
# Files to load
file_to_load = "Resources/missing_grades.csv"

# Read the CSV into a DataFrame
missing_grade_df = pd.read_csv(file_to_load)
missing_grade_df
```

```python
# Drop the NaNs
    # After these rows are dropped, the index of the DataFrame is not automatically reset to 
        # number the rows in consecutive order 
missing_grade_df.dropna()
```

```python
# Fill in the empty rows with "85".
missing_grade_df.fillna(85)
```

```python
# Create a variable missing_grade_df that is equal to the
    # missing_grade_df with the missing values filled with 85
missing_grade_df = missing_grade_df.fillna(85)

# Save previously defined variable of filled DataFrame back to csv file 
missing_grade_df.to_csv('Resources/missing_grades.csv')

# Display the saved dataframe with the filled values
missing_grade_df

# CAN'T HAVE FILE OPEN WHILE SAVING (Running this step)!!!
```
